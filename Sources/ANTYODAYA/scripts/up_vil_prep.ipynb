### Urban shapes

```python
br_urban = gpd.read_file(r'D:\CDL\flood-data-ecosystem-Bihar\Maps\br-ids-drr_shapefile\br_villages_final_filtered.geojson')
br_urban
```

```python
br_urban = br_urban.dropna(subset=['vilnam_soi'])
br_urban = br_urban.loc[br_urban["stcode11"] == "10"]
br_urban["vilnam_soi"] = br_urban["vilnam_soi"].str.upper()
br_urban = br_urban[~br_urban["vilnam_soi"].str.contains("FOREST", na=False)]
br_urban = br_urban[~br_urban["vilnam_soi"].str.contains("RF", na=False)]
br_urban = br_urban[~br_urban["vilnam_soi"].str.contains("RESERVED FOREST", na=False)]
br_urban = br_urban[~br_urban["vilnam_soi"].str.contains(" RF", na=False)]
br_urban = br_urban[~br_urban["vilnam_soi"].str.contains("HILL", na=False)]
br_urban = br_urban[~br_urban["vilnam_soi"].str.contains("R F", na=False)]

br_urban
```

```python
# https://en.wikipedia.org/wiki/List_of_urban_agglomerations_in_Uttar_Pradesh
urban_filter = ['(NPP)','(M Corp.)','(NP)','(OG)','(CB)','(CT)']
filtered_gdf = br_urban[br_urban['vilname11'].str.contains('|'.join(urban_filter), na=False)]
filtered_gdf
```

```python
merged = filtered_gdf.merge(br_vil[['vilcode11']],left_on='vilcode11', right_on='vilcode11',how='left')

merged

```

```python
merged = merged.drop('vilcode11',axis=1)
final_gdf = gpd.GeoDataFrame(merged,geometry='geometry')
final_gdf.to_file(r'D:\CDL\flood-data-ecosystem-Bihar\Maps\br-ids-drr_shapefile\br_urban_final.geojson')
```
